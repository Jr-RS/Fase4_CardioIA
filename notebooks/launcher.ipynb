```python
import os
import sys
import subprocess
from pathlib import Path

REPO_URL = "https://github.com/Jr-RS/Fase4_CardioIA.git"
IN_COLAB = "google.colab" in sys.modules
if IN_COLAB:
    repo_dir = Path("/content/Fase4_CardioIA")
    if not repo_dir.exists():
        subprocess.run(["git", "clone", REPO_URL, str(repo_dir)], check=True)
    os.chdir(repo_dir)
else:
    repo_dir = Path.cwd()
    if not (repo_dir / "requirements.txt").exists():
        raise RuntimeError("Execute este notebook a partir da raiz do repositório ou no Google Colab.")

print(f"Diretório de trabalho: {Path.cwd()}")
subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt"], check=True)
```
